# Helmholtz Decomposition of MITgcm Velocities #

A 2D velocity field can be expressed as
$$ \boldsymbol{u} = \nabla \phi + \hat{\boldsymbol{k}} \cdot \nabla \times \psi  $$ 

Apparently MITgcm can do this decomposition for you using the CG2D solver.

```python
import numpy as np
from MITgcmutils.mds import rdmds
from mitgcmdata import MITgcmmodel
from matplotlib import pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (16,12)
plt.rcParams['image.cmap'] = 'RdBu_r'
plt.rcParams['image.origin'] = 'bottom'
import os
```

```python
base_dir = '/vega/physo/projects/global_tracer/'
vel_dir_orig = os.path.join(base_dir, 'avisoDUACS2014_0.1_degree-msla')
vel_dir_corr = os.path.join(base_dir, 'avisoDUACS2014_0.1_degree-msla-corr')
gcm_dir = os.path.join(base_dir,'GCM/run_correction')
m = MITgcmmodel.ModelInstance(gcm_dir)
```


### Info from Available Diagnostics ##

    48 |PhiVEL  |  1 |    45 |SMR P   MR|m^2/s           |Horizontal Velocity Potential (m^2/s)
    49 |PsiVEL  |  1 |    48 |SZ  P   MR|m.m^2/s         |Horizontal Velocity Stream-Function
    
S = Scalar Diagnostic

M = C-Grid Mass Point

Z = C-Grid Vorticity (Corner) Point

```python
# here is how the numbering of files matches up
# original -> new
# 1        -> 1
# 2        -> 3
# 3        -> 5

n_orig = 10
n_new = 2*n_orig - 1

u0 = np.fromfile(os.path.join(vel_dir_orig, 'uvel.%010d.data' % n_orig), dtype='>f4')
v0 = np.fromfile(os.path.join(vel_dir_orig, 'vvel.%010d.data' % n_orig), dtype='>f4')

phi = m.rdmds('PhiVEL', n_new)
psi = m.rdmds('PsiVEL', n_new)

u0.shape = phi.shape
v0.shape = phi.shape

## non-divergent velocity
# u = - d psi / dy
# v = d psi / dx
u_psi = -(np.roll(psi, -1, axis=-2) - psi) / m.dyg / m.drf
v_psi = (np.roll(psi, -1, axis=-1) - psi) / m.dxg / m.drf

## divergent velocity 
# u = d phi / dx
# v = d phi / dy
# psi is weighted by cell thickness buy phi is not!
u_phi = (phi - np.roll(phi, 1, axis=-1)) / m.dxg # / m.drf
v_phi = (phi - np.roll(phi, 1, axis=-2)) / m.dyg # / m.drf
# do we need to do anything funky with boundary conditions?
# let's try this
maskW = (m.hFacW==0)[np.newaxis,:,:]
maskS = (m.hFacS==0)[np.newaxis,:,:]
u_phi[maskW] = 0
v_phi[maskS] = 0


# alternate way to get non-divergent velocity
u_nd = u0 - u_phi
v_nd = v0 - v_phi

```

## Original Velocity Fields ##

```python
plt.imshow(np.ma.masked_array(u0, m.mask))
plt.clim([-1, 1])
plt.title('Original UVEL')
plt.colorbar(shrink=0.2)

plt.figure()
plt.imshow(np.ma.masked_array(v0, m.mask))
plt.clim([-1, 1])
plt.title('Original VVEL')
plt.colorbar(shrink=0.2)
```

## $\Phi$ and $\Psi$  ##

```python
plt.imshow(phi.squeeze())
plt.title('$\Phi$')
plt.colorbar(shrink=0.2)

plt.figure()
plt.imshow(psi.squeeze())
plt.title('$\Psi$')
plt.colorbar(shrink=0.2)
```

$\Phi$ is nice and smooth. But if you look closely at $\Psi$, you can see the tile outlines. This screws up the derivatives you have to do to recover u and v.

```python
for u, tit in [(u0, 'U Orig'), (u_psi, 'U Psi') , (u_phi, 'U Phi'),
               (u_nd, 'U ND'),
               (u0-u_psi, 'U Orig - U Psi'), (u0 - u_nd, 'U Orig - U ND')]:
    plt.figure()
    plt.imshow(u.squeeze())
    plt.clim([-1, 1])
    plt.xlim([2800,3600])
    plt.ylim([200,500])
    plt.title(tit)
    plt.colorbar(shrink=0.2)
```
